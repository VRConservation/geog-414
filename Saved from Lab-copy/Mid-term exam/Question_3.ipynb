```python
# import ee and geemap define map
import ee 
import geemap
geemap.ee_initialize()
Map = geemap.Map()
Map
```

```python
# Add global temperature data to map along with visParams
dataset = ee.ImageCollection('NOAA/GFS0P25') \
                  .filter(ee.Filter.date('2023-06-28', '2023-06-29'))
temperatureAboveGround = dataset.select('temperature_2m_above_ground')
visParams = {
  'min': -40.0,
  'max': 35.0,
  'palette': ['blue', 'purple', 'cyan', 'green', 'yellow', 'red'],
}
Map.setCenter(-99.46, 39.42, 4)
Map.addLayer(temperatureAboveGround, visParams, 'Temperature Above Ground')
```

```python
# Add states FC
States = ee.FeatureCollection("TIGER/2018/States")
style = {'color': 'black', 'fillColor': '00008000'}
Map.addLayer(States.style(**style), {}, 'USA')
```

```python
# Add temp clip to map turn off global temp layer
Temp_clip = temperatureAboveGround.mean().clip(States)
Map.addLayer(temperatureAboveGround, visParams, 'Temperature Above Ground', False)
Map.addLayer(Temp_clip, visParams, 'Temperature Above Ground USA')
Map
```

```python
# Export the image to CSV, specifying scale and region.
temp_stats = 'temp_stats.csv'
geemap.zonal_stats(
    Temp_clip, States, temp_stats, statistics_type='MAXIMUM', scale=1000, return_fc=False)
```

```python
#Display to 10 rows of the attribute table, reverse order to see highest temperature
geemap.csv_to_df(temp_stats).sort_values('max', ascending=False).head(10)
# Arizona
```
