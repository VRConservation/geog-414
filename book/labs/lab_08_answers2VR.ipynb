### Question 3
Write a SQL query to find all subway stations in the `nyc_subway_stations` dataset that are located in the borough of Manhattan.

```python
%%sql

SELECT * FROM stations WHERE BOROUGH='Manhattan'
```

### Question 4
Write a SQL query to list the subway stations in the `nyc_subway_stations` dataset in alphabetical order by their names.

```python
%%sql 

SELECT * FROM stations ORDER BY NAME;
```

### Question 5
Write a SQL query to find the distinct boroughs represented in the `nyc_subway_stations` dataset.

```python
%%sql

SELECT DISTINCT BOROUGH FROM stations;
```

### Question 6
count the number of subway stations in each borough in the `nyc_subway_stations` dataset.

```python
%%sql

SELECT BOROUGH, COUNT (*) AS NAME
FROM stations
GROUP BY BOROUGH;
```

### Question 7
list the number of subway stations in each borough, sorted in descending order by the count.

```python
%%sql

SELECT BOROUGH, NAME FROM stations
GROUP BY BOROUGH, NAME
ORDER BY NAME DESC;
```
