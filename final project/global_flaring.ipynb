```python
# Import libraries
import duckdb
import leafmap
import pandas as pd
```

```python
# Connect to duckdb and install extensions
con = duckdb.connect()
con.install_extension("httpfs")
con.load_extension("httpfs")
con.install_extension("spatial")
con.load_extension("spatial")
```

```python
# Download and the 2021 and 2017 gas flare datasets 
# Downloaded to file since and converted to csv since it wouldn't read xlsx at the link above dropped to the current folder
con.read_csv('gas21.csv')
```

```python
df = pd.read_csv('gas21.csv')
df
```

```python
# Convert to geodataframe
from shapely.geometry import Point
import geopandas as gpd

gdf = gpd.GeoDataFrame(df, geometry=gpd.points_from_xy(df.Longitude, df.Latitude))

# Set the GeoDataFrame's coordinate reference system
gdf.crs = 'EPSG:4326'
gdf
```

```python
# Visualize using lonboard
leafmap.view_vector(gdf, get_radius=20000, get_fill_color='red', opacity=0.1, get_line_color='black', get_line_width=1)
```
