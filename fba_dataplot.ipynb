```python
# import packages
import seaborn as sns
import pandas as pd
from matplotlib import pyplot as plt
```

```python
# Load data and view df. (When I transferred this to onedrive it wouldn't read it)
# So left it in the downloads folder
fba = pd.read_csv('C:/Users/vance/Downloads/fba.csv')
fba
```

```python
# Plot pipeline counts
fba.groupby('Pipeline').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
# scatterplot timeline with legend in upper left

sns.scatterplot(data=fba, x='Date_1st_Contact', y='Type', hue='Pipeline', palette='Dark2')
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0)
plt.show()
```
